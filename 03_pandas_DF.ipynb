# Task Write a function that take two teams names and writen which team wins how many match

```python
def Tracker(Team1,Team2):
    
    Team1_wins = Ipl[Ipl['WinningTeam'] == Team1 ].shape[0]
    Team2_wins = Ipl[Ipl['WinningTeam'] == Team2 ].shape[0]

    print("Number of matches won by ", Team1,"is", Team1_wins)
    print("Number of matches won by ", Team2,"is", Team2_wins)

Tracker('Gujarat Titans','Chennai Super Kings')
```
