### Pivot Table

The pivot table takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data.

# Pandas Pivot Table - Complete Guide

## What is a Pivot Table?

A **Pivot Table** is a powerful feature in Pandas used to **summarize, organize, and analyze data**. It groups data based on one or more columns and applies aggregation functions such as **sum, mean, count, min, max**, etc.

It is similar to the Pivot Table feature in Microsoft Excel.

### Why Use a Pivot Table?

Pivot tables help you answer questions like:

- What is the total sales for each product?
- What is the average salary in each department?
- How many students are in each class?
- Which city has the highest number of customers?

Instead of reading thousands of rows, a pivot table gives a quick summary.

---

# Syntax

```python
pd.pivot_table(
    data,
    values=None,
    index=None,
    columns=None,
    aggfunc='mean',
    fill_value=None,
    margins=False,
    margins_name='All',
    dropna=True,
    observed=False,
    sort=True
)
```

# Parameters

The DataFrame you want to summarize.

## 2. values

The column(s) you want to calculate.

Single column


Multiple columns


If `values` is not given, Pandas tries to aggregate all numeric columns.

# Common Aggregation Functions

| Function | Description |
|----------|-------------|
| mean | Average value |
| sum | Total value |
| count | Count non-missing values |
| size | Count all rows |
| min | Smallest value |
| max | Largest value |
| median | Middle value |
| std | Standard deviation |
| var | Variance |
| first | First value |
| last | Last value |
| nunique | Count unique values |
| prod | Product of values |

---

# Pivot Table vs Pivot

| Feature | pivot() | pivot_table() |
|----------|----------|---------------|
| Aggregation | ❌ No | ✅ Yes |
| Handles duplicate values | ❌ No | ✅ Yes |
| Multiple aggregation functions | ❌ No | ✅ Yes |
| Replace missing values | ❌ No | ✅ Yes |
| Grand totals | ❌ No | ✅ Yes |

---

# Pivot Table vs GroupBy

| Feature | groupby() | pivot_table() |
|----------|-----------|---------------|
| Data summary | ✅ | ✅ |
| Row grouping | ✅ | ✅ |
| Column grouping | Manual | Built-in |
| Spreadsheet-style output | ❌ | ✅ |
| Grand totals | Manual | Built-in |

# Advantages of Pivot Table

- Easy to summarize large datasets.
- Can group data by one or more columns.
- Supports many aggregation functions.
- Can display row and column summaries.
- Can replace missing values.
- Similar to Excel Pivot Tables.
- Easy to read and understand.

---

# Limitations

- Works best with structured data.
- Only aggregation functions can be applied.
- Very large datasets may take more time to process.

# Best Practices

- Use `pivot_table()` when your data contains duplicate values.
- Use `pivot()` only when every row-column combination is unique.
- Choose the correct `aggfunc` for your analysis.
- Use `fill_value` to replace missing values.
- Use `margins=True` to add totals.
- Use multiple indexes and columns for detailed reports.
- Use dictionaries in `aggfunc` when different columns need different calculations.

# Pivot syntax  pivot_table(index=,columns=,values=,aggfunc) default value of aggfunc is mean

````python
# using pivot 
df.pivot_table(index='sex',columns='smoker',values='total_bill')
````

````python
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='std')
````

````python
# multidimensional

df.pivot_table(
    index=['sex', 'smoker'],
    columns=['day', 'time'],
    values='total_bill'
)
````

````python
df.pivot_table(
    index=['sex','smoker'],
    columns=['day','time'],
    aggfunc={'size':'mean',
             'tip':'max',
             'total_bill':'sum'}
)
````

````python
#margins
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum',margins=True) # margin for sum
````
